```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import OneHotEncoder, RobustScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import category_encoders as ce
```

```python
df = pd.read_csv("../datasets/dataframe_for_preprocessing.csv")
df.head()
```

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
print(df.duplicated(keep=False).sum())
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
df.describe().T
```

```python
constant_cols = [col for col in df.columns if df[col].min()==df[col].max()]
df.drop(columns=constant_cols,inplace=True)
```

```python
df.shape
```

```python
df.describe(include="object").T
```

```python
low_cardinality = ["protocol_type"]
medium_cardinality = ['flag']
high_cardinality = ['service']
```

```python
def get_numerical(df):
    numerical_cols = []
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        if df[col].nunique(dropna=True) > 2:
            numerical_cols.append(col)
    return numerical_cols

def get_binary(df):
    binary_cols = []
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        if df[col].nunique(dropna=True) == 2:
            binary_cols.append(col)
    return binary_cols

def convert_to_folat(X):
    return X.astype(float)
```

```python
numerical_cols = get_numerical(df)
binary_cols = get_binary(df)
```

```python
log_pipeline = Pipeline(
    steps=[
        ('to_float', FunctionTransformer(convert_to_folat, validate=False)),
        ('log', FunctionTransformer(np.log1p, validate=True)),
        ('scaler', RobustScaler())
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        
        ('num_log_scale', log_pipeline, numerical_cols),

        ('binary_pass','passthrough',binary_cols),
        ('low_ohe', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'),
         low_cardinality),

        ('medium_ohe', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'),
         medium_cardinality),

        ('high_binary', ce.BinaryEncoder(), high_cardinality)
    ],
    remainder='drop' 
)
```

```python
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')
```

```python
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
X_train_final = preprocessor.fit_transform(X_train)
X_test_final = preprocessor.transform(X_test)
```

```python
iso_forest = IsolationForest(n_estimators=500, contamination=0.1, random_state=42)
```

```python
iso_forest.fit(X_train_final)
```

```python
if_scores = iso_forest.decision_function(X_test_final)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(if_scores, bins=50, kde=True, color='darkorange')
plt.axvline(x=0, color='red', linestyle='--') 
plt.title("Distribution of Isolation Forest Decision Scores")
plt.xlabel("Anomaly Score (Negative = More Anomalous)")
plt.show()
```

```python
y_pred_iso = iso_forest.predict(X_test_final)
print(f"iso detected {sum(y_pred_iso == -1)} anamalies.")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_test_final)
```

```python
normal_points = X_pca[y_pred_iso == 1]
anomaly_points = X_pca[y_pred_iso == -1]
plt.figure(figsize=(10, 7))
plt.scatter(normal_points[:, 0], normal_points[:, 1], 
            c='royalblue', label='Inside Bubble (Normal)', alpha=0.5, s=15)
plt.scatter(anomaly_points[:, 0], anomaly_points[:, 1], 
            c='orange', label='Outside Bubble (Anomaly)', alpha=0.8, s=20, edgecolors='black')

plt.title("Isolation Forest: The 'Normal Behavior' Boundary")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```

```python
print(f"Total Variance Explained by 2 components: {pca.explained_variance_ratio_.sum():.2%}")
```

```python
import joblib
from sklearn.pipeline import Pipeline

final_model = Pipeline(steps=[
    ('Preprocessor',preprocessor),
    ('model', iso_forest)
])
joblib.dump(final_model, 'iso_forest_network_pipeline.joblib')
print("Success: Preprocessor and Isolation Forest saved as 'iso_forest_network_pipeline.joblib'")
```
